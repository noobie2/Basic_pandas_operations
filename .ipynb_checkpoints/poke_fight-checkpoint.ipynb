#### A simple fighting function which will check total attributes for the pokemon and return who won.

```python
def fight(pok1,pok2):
    po1 = df1[df1.Name==pok1]
    po2 = df1[df1.Name==pok2]
    if po1.loc[:,'HP':'Speed'].sum().sum() > po2.loc[:,'HP':'Speed'].sum().sum():
        return(f'{pok1} won!!!')
    else:
        return(f'{pok2} won!!!')
fight('Weedle','Kakuna')
```

#### Can be used to filter pokemon on basis of type,gen and legendary status.

```python
def filter_pok():
    print('Specify filtering categories:\n1: Pokemon Type - ',sorted(df1['Type 1'].unique()),
          '\n2: Pokemon Generation - 1 to 6 Gen'
          '\n3:Is Pokemon Legendary? - Y/N')
    pok_type,pok_gen,pok_leg = input().split()
    pok_leg = True if pok_leg.lower() == 'y' else False
    return df1.loc[((df1['Type 1'].str.lower() == pok_type.lower()) | (df1['Type 2'].str.lower() == pok_type.lower()))
                   & (df1['Generation'] == int(pok_gen))
                   & (df1['Legendary'] == pok_leg)]
```

#### Return a dataframe of legendary pokemons.

```python
def get_legend():
    return df1[df1['Legendary']]
```
